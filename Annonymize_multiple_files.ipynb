```python
#input your key columns and columns you want to anonymize
if c == 1:
  df_dup = df1.copy()
  col_names = df1.columns.tolist()
  for i in range(len(col_names)):
    print(i, col_names[i])
  k = eval (input("Enter the key column: "))
  user_col = []
  n=1
  print("Enter negative value to exit.")
  while n>=0:
    n = eval (input("Enter the column number that you want to annonymize: "))
    if n>=0:
      user_col.append(n)
```

```python
#anonymization happens here 
S = 5  # number of characters in the string.
dict_key={}
key_values = df1.iloc[:, k].tolist()
rand_keys = []
for i in range(len(df1)):
  ran = ''.join(random.choices(string.ascii_uppercase + string.digits, k = S))
  rand_keys.append(ran)
  list_1 = []
  for j in range(len(user_col)):
    list_1.append(df1.iloc[i, user_col[j]]) 
  dict_key[ran]= list_1
print(dict_key)
```

```python
#write to csv files. 'Annony.csv' contains csv where the columns to be annonymized is replaced with anonymous key. 'Key.csv' contains file which contains the anonymous key along with the columns which were dropped from original file.
df_dup.drop(df_dup.columns[user_col], axis = 1, inplace = True)
df_dup['Annonymized Key'] = rand_keys
df_dup.to_csv("Annony.csv")
user_col.append(k)
df_key = df1.iloc[:, user_col]
df_key['Annonymized Key'] = rand_keys
df_key.to_csv("Key.csv")
```
