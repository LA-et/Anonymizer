This program takes the input of a file and print the names of all the columns. The user can select the column number and that particular column will then be annonymized and the program will then return two csv files one containing the key and value and other the annonimized file.

```python
import numpy as np
import pandas as pd
import string    
import random
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
#Enter the name of the file
file_name=input("Enter the file name with .csv extension")
```

```python
import io
df2 = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

```python
df2
```

```python
col_name=list((df2.columns))
```

```python
dict_col_name={}
for i in range(1,len(list(df2.columns))+1):
  dict_col_name[i]=col_name[i-1]
```

```python
dict_col_name
```

```python
#enter the column numbers to be annonimized seperated by commas without space
col_num=input("enter the column numbers to be annonimized seperated by commas without space")
```

```python
col_name_list= [int(i) for i in col_num.split(",") if i.isdigit()]
```

```python
col_name_list
```

```python
dict_col_name[col_name_list[0]]
```

```python
S = 5  # number of characters in the string.

key_value=pd.DataFrame({})
for j in col_name_list:
  dict_key={}
  for i in df2[dict_col_name[j]]:
    ran = ''.join(random.choices(string.ascii_uppercase + string.digits, k = S))
    dict_key[ran]= i
  key_value[dict_col_name[j]]=list(df2[dict_col_name[j]])
  key_value[dict_col_name[j]+str("_key")]=list(dict_key.keys())
  df2[dict_col_name[j]]=list(dict_key.keys())
  
```

```python
key_value.to_csv("key_value.csv")
```
